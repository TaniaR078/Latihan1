pembaharuan variabel pack yang memiliki category_name CREAM LIQUEURS diubah menjadi 40

```python
# update database variabel pack yang memiliki category_name CREAM LIQUEURS diubah menjadi 40
update_query = """ UPDATE sales SET pack='40' WHERE category_name='CREAM LIQUEURS'
"""

conn.execute(update_query)
```

```python
# memanggil data yang sudah diupdate
pd.read_sql_query("SELECT Category_name,pack FROM sales WHERE category_name= 'CREAM LIQUEURS'", conn)
```

```python
# Update pada table sales kolom sale_dollars yang memiliki  item_descriptions TITOS HANDMADE VODKA diubah menjadi 400
update_query = """ UPDATE sales SET sale_dollars='400' WHERE item_description='TITOS HANDMADE VODKA'
"""

conn.execute(update_query)
```
