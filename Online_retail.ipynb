# NO 1

Persentase growth pada total sales pada setiap bulannya

Expected Output : periode (year/month), total_sales, prev_month_sales, sales_growth

```python
# Persentase growth pada total sales pada setiap bulannya
query = """SELECT
    STRFTIME('%Y/%m', Order_Date) AS periode,
    SUM(s.quantity * p.retail_price * (1 - c.loyalty_discount)) AS total_sales,
    LAG(SUM(s.quantity * p.retail_price * (1 - c.loyalty_discount))) OVER (ORDER BY STRFTIME('%Y/%m', Order_Date)) AS prev_month_sales,
    ((SUM(s.quantity * p.retail_price * (1 - c.loyalty_discount)) - LAG(SUM(s.quantity * p.retail_price * (1 - c.loyalty_discount))) OVER (ORDER BY STRFTIME('%Y/%m', Order_Date))) / LAG(SUM(s.quantity * p.retail_price * (1 - c.loyalty_discount))) OVER (ORDER BY STRFTIME('%Y/%m', Order_Date))) * 100
    AS sales_growth
FROM
    sales s
JOIN
    product p ON s.item_id = p.item_id
JOIN
    customer c ON s.customerID = c.customerID
GROUP BY periode"""
pd.read_sql_query(query, conn)
```

# NO.2

Kategori pelanggan berdasarkan spending (jumlah yang dibayarkan ketika membeli barang) yang dibagi menjadi 3 kategori yaitu

1. High_spender (>500), Mid_spender (300-500) dan Low_spender (<300)
2. Tentukan juga jumlah customer pada masing-masing category, rata-rata jumlah barang yang dibeli dan rata-rata spending
3. Pengkategorian dibuat tanpa memasukan data pelanggan yang berada di Wyomin, Idaho dan Alaska
4. Expected Output : cust_cat (customer category), total_user, avg_quantity, dan avg_spending

```python
# Kategori pelanggan berdasarkan spending (jumlah yang dibayarkan ketika membeli barang) yang dibagi menjadi 3 kategori yaitu High_spender, Mid_spender, dan Low_spender
query ="""WITH CustomerSpending AS (
    SELECT
        c.Customerid,
        SUM(s.quantity * p.retail_price * (1 - c.loyalty_discount)) AS total_spending
    FROM
        sales s
    JOIN
        customer c ON c.customerid = s.customerid
    JOIN
        product p ON s.item_id = p.item_id
    WHERE
        s.shipping_state NOT IN ('Wyomin', 'Idaho', 'Alaska')
    GROUP BY
        c.customerid
),
CustomerCategories AS (
    SELECT
        Customerid,
        total_spending,
        CASE
            WHEN total_spending > 500 THEN 'High_spender'
            WHEN total_spending >= 300 AND total_spending <= 500 THEN 'Mid_spender'
            WHEN total_spending < 300 THEN 'Low_spender'
        END AS cust_cat
    FROM
        CustomerSpending
)
SELECT
    cust_cat,
    COUNT(s.customerid) AS total_user,
    AVG(s.quantity) AS avg_quantity,
    AVG(total_spending) AS avg_spending
FROM
    sales s
JOIN
     CustomerCategories cc ON cc.customerid = s.customerid
GROUP BY
    cust_cat
ORDER BY
    cust_cat;
"""
pd.read_sql_query(query, conn)
```

Ranking setiap produk pada tahun 2004 berdasarkan rata-rata rating. Buat dalam 1 tabel yang berisikan 5 produk yang memiliki rating tertinggi dan 5 produk yang memiliki rating terendah

1. Tidak boleh ada produk yang memiliki ranking yang sama
2. Untuk membuat ke dalam 1 tabel, bisa menggunakan metode yang ada di Link berikut
3. Expected Output : Year, Product, avg_rating, Ranking

```python
# Ranking setiap produk pada tahun 2004 berdasarkan rata-rata rating
query ="""WITH RankedProducts AS (
        SELECT
            p.Description AS Product,
            AVG(s.Rating) AS avg_rating,
            RANK() OVER (ORDER BY AVG(s.Rating) DESC) AS RankDesc,
            DENSE_RANK() OVER (ORDER BY AVG(s.Rating) DESC) AS DenseRankDesc,
            ROW_NUMBER() OVER (ORDER BY AVG(s.Rating) DESC) AS RowNum
        FROM
            Sales s
        INNER JOIN
            Product p USING(Item_id)
        WHERE
            strftime('%Y', s.Order_Date) = '2004'
        GROUP BY
            s.Item_id, p.Description
        HAVING
            COUNT(s.Rating) > 0
    )
    SELECT
        2004 AS Year,
        Product,
        avg_rating,
        RowNum AS Ranking
    FROM
        RankedProducts
    WHERE
        RowNum <= 5

    UNION

    SELECT
        2004 AS Year,
        Product,
        avg_rating,
        RowNum AS Ranking
    FROM
        RankedProducts
    WHERE
        DenseRankDesc > 5
    ORDER BY
        avg_rating DESC, Ranking ASC;"""
pd.read_sql_query(query, conn)
```
